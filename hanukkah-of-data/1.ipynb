```python
def c_to_digit(c: str):
    if c == " ":
        return ""
    if c <= "c":
        return "2"
    if c <= "f":
        return "3"
    if c <= "i":
        return "4"
    if c <= "l":
        return "5"
    if c <= "o":
        return "6"
    if c <= "s":
        return "7"
    if c <= "v":
        return "8"
    return "9"


def s_to_phone(s: str):
    s = s.lower().split(" ")[1]
    # s = s.lower()
    return "".join(map(c_to_digit, s))
```

```python
customers_df["phone_from_name"] = customers_df["name"].apply(s_to_phone)
```

```python
customers_df["cleaned_phone"] = customers_df["phone"].str.replace("-", "")
```
